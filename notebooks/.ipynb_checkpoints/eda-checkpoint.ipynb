For each headline, it calculates a sentiment polarity score, which measures the emotional tone of the text on a scale from -1 (very negative) to 1 (very positive).

Topic Modeling in NLP is an unsupervised machine learning technique used to discover abstract topics within a collection of documents. It identifies patterns of word co-occurrence across documents and groups words with similar meanings into topics. Each document is represented as a mixture of these topics, and each topic is characterized by a distribution of words.

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK stopwords and punkt tokenizer if not already available
nltk.download('stopwords')
nltk.download('punkt')

# Ensure the correct path explicitly
nltk.data.path.append('C:\\Users\\Dagi\\AppData\\Roaming\\nltk_data')
```

Clean the text data by removing special characters, numbers, and converting to lowercase.

```python
# Assuming 'df' is your DataFrame containing the 'headline' column
df['clean_headline'] = df['headline'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x.lower()))

# Tokenization and removing stopwords
stop_words = set(stopwords.words('english'))
df['tokens'] = df['clean_headline'].apply(lambda x: [word for word in word_tokenize(x) if word not in stop_words])
```

Convert the text data into a document-term matrix using CountVectorizer or TfidfVectorizer.

```python
# Convert tokens back to a single string per document
df['cleaned_text'] = df['tokens'].apply(lambda x: ' '.join(x))

vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(df['cleaned_text'])
df.head()
```

```python
from collections import Counter
import re

# Combine all texts in the 'cleaned_text' column into one large string
all_text = ' '.join(df['cleaned_text'].tolist())

# Tokenize the text into words using regex, converting to lowercase to count consistently
words = re.findall(r'\b\w+\b', all_text.lower())

# Count the frequency of each word
word_counts = Counter(words)

# Get the top 10 most common words
top_10_common_words = word_counts.most_common(10)

# Print the results
print("Top 10 Common Words:")
for word, count in top_10_common_words:
    print(f"{word}: {count}")
```

```python
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(df['cleaned_text'])

# Display the result
print(df.head())
print(dtm.toarray())

# Get the top 10 common words (as you already did)
all_text = ' '.join(df['cleaned_text'].tolist())
words = re.findall(r'\b\w+\b', all_text.lower())
word_counts = Counter(words)
top_10_words = [word for word, count in word_counts.most_common(10)]

# Initialize a dictionary to store word counts per year
word_freq_per_year = {word: [] for word in top_10_words}
years = sorted(df['year'].unique())

# Calculate the frequency of each top word per year
for year in years:
    yearly_text = ' '.join(df[df['year'] == year]['cleaned_text'].tolist())
    yearly_words = re.findall(r'\b\w+\b', yearly_text.lower())
    yearly_word_counts = Counter(yearly_words)
    for word in top_10_words:
        word_freq_per_year[word].append(yearly_word_counts[word])

# Plot the word frequencies over the years
plt.figure(figsize=(10, 6))
for word in top_10_words:
    plt.plot(years, word_freq_per_year[word], label=word)

plt.title('Top 10 Common Words Repetition Over the Years')
plt.xlabel('Year')
plt.ylabel('Word Frequency')
plt.legend(title="Words")
plt.grid(True)
plt.savefig('most_common_words_over_the_years.png')
plt.show()
```

Latent Dirichlet Allocation (LDA): Extract topics from the document-term matrix.

```python
lda = LatentDirichletAllocation(n_components=5, random_state=42)  
lda.fit(dtm)
```

```python
# Display the top words per topic
for index, topic in enumerate(lda.components_):
    print(f"Topic {index + 1}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
```
